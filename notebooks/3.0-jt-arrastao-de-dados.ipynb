## Passo 3 - Coletando todos os dados

```python
from src.data.make_dataset import ColetorLinksPizza, ColetorDadosEpisodio, parse_pagina
```

Usando o coletor de links para juntar os links dos episódios

```python
c = ColetorLinksPizza()
c.arrasta()
```

```python
c.links.reverse()
episodios = c.links
```

```python
dados = []
for indice, url in enumerate(episodios):
    c = ColetorDadosEpisodio(url)
    c.arrasta()
    dados_episodio = c.serializado()
    dados_episodio['id'] = indice + 1
    dados.append(dados_episodio)
```

___
## Montando um dataset

```python
import pandas as pd
```

```python
df = pd.DataFrame(dados)
df
```

___
## Salvando os dados

```python
df.to_csv('../data/raw/2019-08-11-dados-do-pizza.csv', index=False)
```
